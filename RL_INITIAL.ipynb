```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
#this is the class for the simplest environment where the agent can move from state 0 to state 4 .
#in init function, the different states are defined and the start state is set to be 2
#in reset function, the state value is set to 2
#in step function, when state is greater than 4 or less than 0, the "state value", "reward", "Termination_flag" with the corresponding values is returned.
#else the current "state", "reward" and "termination flag" is returned
class SimpleEnv:
    def __init__(self):
        self.states = [0,1,2,3,4]
        self.start_state = 2
        self.state = self.start_state

    def reset(self):
        self.state = self.start_state
        return self.state

    def step(self, action):
        self.state += action
        if self.state>=4:
            return 4, 1, True                      #"state", "reward", "Termination_flag"
        elif self.state <=0:
            return 0, -1, True                     #"state", "reward", "Termination_flag"
        else:
            return self.state, 0, False            #"state", "reward", "Termination_flag"
```

```python
#this is the class to define the random agent which randomly moves to +1(right) and -1(left) direction
#policy function is responsible to generate random numbers either -1 or +1

class RandomAgent:
    def __init__(self):
        self.total_reward = 0

    def policy(self, state):
        return random.choice([-1,1])  #returns randomly either -1 or 1
```

```python
#Creating objects
env = SimpleEnv()
agent = RandomAgent()
```

```python
#Randomly running the agent in the environment until it reaches the terminal state:
#here in this loop, no training is happening, just an agent moving randomly.
#when the terminal state i.3 either 0 or 4 hits, the function, step of class 
EPISODES = 10
for ep in range(EPISODES):
    state = env.reset()
    done = False
    ep_reward = 0

    print(f"no. of epochs is {ep+1}")
    while not done:
        action = agent.policy(state)
        next_state, reward, done = env.step(action)
        ep_reward += reward
        print(f"state:{state} -> Action: {action:+} -> Next: {next_state}, Reward: {reward}")
        state = next_state

    print(f" Episode {ep+1} ended. Total reward: {ep_reward}\n")
```
