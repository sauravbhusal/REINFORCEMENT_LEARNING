```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
class SimpleEnv:
    def __init__(self):
        self.states = [0,1,2,3,4]
        self.start_state = 2
        self.state = self.start_state

    def reset(self):
        self.state = self.start_state
        return self.state

    def step(self, action):
        self.state += action
        if self.state>=4:
            return 4, 1, True
        elif self.state <=0:
            return 0, -1, True
        else:
            return self.state, 0, False
```

```python
class RandomAgent:
    def __init__(self):
        self.total_reward = 0

    def policy(self, state):
        return random.choice([-1,1])  #returns randomly either -1 or 1
```

```python
env = SimpleEnv()
agent = RandomAgent()
```

```python
EPISODES = 10
for ep in range(EPISODES):
    state = env.reset()
    done = False
    ep_reward = 0

    print(f"no. of epochs is {ep+1}")
    while not done:
        action = agent.policy(state)
        next_state, reward, done = env.step(action)
        ep_reward += reward
        print(f"state:{state} -> Action: {action:+} -> Next: {next_state}, Reward: {reward}")
        state = next_state

    print(f" Episode {ep+1} ended. Total reward: {ep_reward}\n")
```
